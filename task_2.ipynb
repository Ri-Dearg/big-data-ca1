### Connect to MongoDB Atlas
My URI is utilised to connect to the DB, then the connection is verified.

```python
# Import relevant libraries for the connection and using secrets.
from google.colab import userdata
from pymongo.mongo_client import MongoClient
from pymongo.server_api import ServerApi

# The URI is loaded from Colab Secrets (the key icon in the left sidebar).
uri = userdata.get('MONGO_URI')

# Create a new client and connect to the Atlas cluster.
client = MongoClient(uri, server_api=ServerApi('1'))

# Ping the deployment to confirm a successful connection.
try:
    client.admin.command('ping')
    print('Pinged your deployment. You successfully connected to MongoDB!')
except Exception as e:
    print(e)
```

### Create the `members` Database and Insert Data

The data below models each person as a document.  
Favourite sports are stored as a list to allow multiple values per person  
Address is stored as an nested document, as this is how it is represented in the table.

```python
# Define the member data as a list of Python dictionaries.
# Each dictionary is one MongoDB document.
members_data = [
    {
        'first_name': 'Carol',
        'last_name': 'Smith',
        'age': 50,
        'favourite_sports': ['Tennis', 'Badminton'],
        'address': {'city': 'Dublin', 'street': 'Abbey Rd.', 'number': 1},
    },
    {
        'first_name': 'Richard',
        'last_name': 'Miller',
        'age': 30.5,
        'favourite_sports': ['Tennis'],
        'address': {'city': 'Cork', 'street': 'Temple St.'},
    },
    {
        'first_name': 'Thomas',
        'last_name': 'Garcia',
        'age': 55,
        'favourite_sports': ['Football'],
        'address': {'city': 'Galway', 'street': 'Henry St.', 'number': 3},
    },
    {
        'first_name': 'Eugenia',
        'last_name': 'Jones',
        'age': 22,
        'favourite_sports': ['Handball', 'Basketball'],
        'address': {'city': 'Belfast', 'street': 'Oakley Rd.'},
    },
    {
        'first_name': 'Andrew',
        'last_name': 'Hernandez',
        'age': 18,
        'favourite_sports': ['Basketball'],
        'address': {'city': 'Dublin', 'number': 4},
    },
]
```

```python
try: # Create/Select the 'members' database and 'people' collection.
  # If the db is not present, it will be created.
  db = client.members
  people = db.people

  # Insert all people into the database..
  result = people.insert_many(members_data)

except Exception as e:
    print(f"An error occurred: {e}")

# Confirm the people are in the database.
for person in people.find({}):
      print(person)
```

### Q1: Display all information about each person except for their address.

```python
# Use find() with an empty query {} to return all documents.
# Use {"address": 0} exclude the address field from the output.
results_1 = people.find({}, {'address': 0})

# Show results
for person in results_1:
    print(person)
```

### Q2: What age is Carol?

```python
# Use find_one() to return a single document.
# Filter by first_name = 'Carol' and show only the age field.
# _id is excluded with 0 for cleaner output.
result_2 = people.find_one({'first_name': 'Carol'}, {'age': 1, '_id': 0})

# Print answer
print(f'Carol is {result_2["age"]}.')
result_2
```

### Q3: Display the age of all records in descending order. Also display their first and last name.

```python
# Use find() to select only first_name, last_name and age.
# sort("age", -1) sorts by age in descending order.
results_3 = people.find({}, {'first_name': 1, 'last_name': 1, 'age': 1, '_id': 0}
                        ).sort('age', -1)

# Show results
for person in results_3:
    print(person)
```

### Q4: Display the name and surname of everyone who has tennis as their favourite sport.

```python
# Query the favourite sports list directly for Tennis.
# Return only first_name and last_name.
results_4 = people.find(
    {'favourite_sports': 'Tennis'}, {'first_name': 1, 'last_name': 1, '_id': 0}
)

# Show results
for person in results_4:
    print(person)
```

### Q5: Count the number of members who have tennis as their favourite sport.

```python
# Use count_documents() to return the number of people
# with Tennis in favourite_sports
result_5 = people.count_documents({'favourite_sports': 'Tennis'})

result_5
```

### Q6: Display first name, last name, and age of all people older than 30 years old.

```python
# Use $gt to select numeric values greatr than the input value.
# We filter documents where age > 30 and show the three requested fields.
results_6 = people.find(
    {'age': {'$gt': 30}}, {'first_name': 1, 'last_name': 1, 'age': 1, '_id': 0}
)

# Show results
for person in results_6:
    print(person)
```
