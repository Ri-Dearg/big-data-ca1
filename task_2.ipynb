### Create the `members` Database and Insert Data

The data below models each person as a document.  
Favourite sports are stored as a list to allow multiple values per person  
Address is stored as an nested document, as this is how it is represented in the table.

```python
# Define the member data as a list of Python dictionaries.
# Each dictionary is one MongoDB document.
members_data = [
    {
        'first_name': 'Carol',
        'last_name': 'Smith',
        'age': 50,
        'favourite_sports': ['Tennis', 'Badminton'],
        'address': {'city': 'Dublin', 'street': 'Abbey Rd.', 'number': 1},
    },
    {
        'first_name': 'Richard',
        'last_name': 'Miller',
        'age': 30.5,
        'favourite_sports': ['Tennis'],
        'address': {'city': 'Cork', 'street': 'Temple St.'},
    },
    {
        'first_name': 'Thomas',
        'last_name': 'Garcia',
        'age': 55,
        'favourite_sports': ['Football'],
        'address': {'city': 'Galway', 'street': 'Henry St.', 'number': 3},
    },
    {
        'first_name': 'Eugenia',
        'last_name': 'Jones',
        'age': 22,
        'favourite_sports': ['Handball', 'Basketball'],
        'address': {'city': 'Belfast', 'street': 'Oakley Rd.'},
    },
    {
        'first_name': 'Andrew',
        'last_name': 'Hernandez',
        'age': 18,
        'favourite_sports': ['Basketball'],
        'address': {'city': 'Dublin', 'number': 4},
    },
]
```

```python
try: # Create/Select the 'members' database and 'people' collection.
  # If the db is not present, it will be created.
  db = client.members
  people = db.people

  # Drop contents so dupplicates are not added on new runs.
  people.drop()

  # Insert all people into the database..
  result = people.insert_many(members_data)

except Exception as e:
    print(f"An error occurred: {e}")

# Confirm the people are in the database.
for person in people.find({}):
      print(person)
```

### Q1: Display all information about each person except for their address.

```python
# Use find() with an empty query {} to return all documents.
# Use {"address": 0} exclude the address field from the output.
results_1 = people.find({}, {'address': 0})

# Show results
for person in results_1:
    print(person)
```

### Q2: What age is Carol?

```python
# Use find_one() to return a single document.
# Filter by first_name = 'Carol' and show only the age field.
# _id is excluded with 0 for cleaner output.
result_2 = people.find_one({'first_name': 'Carol'}, {'age': 1, '_id': 0})

# Print answer
print(f'Carol is {result_2["age"]}.')
result_2
```

### Q3: Display the age of all records in descending order. Also display their first and last name.

```python
# Use find() to select only first_name, last_name and age.
# sort("age", -1) sorts by age in descending order.
results_3 = people.find({}, {'first_name': 1, 'last_name': 1, 'age': 1, '_id': 0}
                        ).sort('age', -1)

# Show results
for person in results_3:
    print(person)
```

### Q4: Display the name and surname of everyone who has tennis as their favourite sport.

```python
# Query the favourite sports list directly for Tennis.
# Return only first_name and last_name.
results_4 = people.find(
    {'favourite_sports': 'Tennis'}, {'first_name': 1, 'last_name': 1, '_id': 0}
)

# Show results
for person in results_4:
    print(person)
```

### Q5: Count the number of members who have tennis as their favourite sport.

```python
# Use count_documents() to return the number of people
# with Tennis in favourite_sports
result_5 = people.count_documents({'favourite_sports': 'Tennis'})

# Show Results.
print(result_5)
print(f'{result_5} people have tennis as their favourite sport')
```

### Q6: Display first name, last name, and age of all people older than 30 years old.

```python
# Use $gt to select numeric values greatr than the input value.
# We filter documents where age > 30 and show the three requested fields.
results_6 = people.find(
    {'age': {'$gt': 30}}, {'first_name': 1, 'last_name': 1, 'age': 1, '_id': 0}
)

# Show results
for person in results_6:
    print(person)
```

### Q7: Retrieve the name of all members whose first name contains the letter a.

```python
# Use $regex to match patterns in MongoDB.
# 'a' matches any first_name containing the letter 'a'.
# The 'i' flag makes the search case-insensitive.
results_7 = people.find(
    {'first_name': {'$regex': 'a', '$options': 'i'}},
    {'first_name': 1, 'last_name': 1, '_id': 0},
)

# Show results
for person in results_7:
    print(person)

# All people have 'a' in their name if it's checked case-insensitive.
# Otherwise, Andrew would be excluded.
```

### Q8: Calculate the average age of everyone in the database.

```python
# Use aggregations for calculations across documents.
# $group with _id: None groups all documents together into a one result.
# $avg calculates the average 'age' across all documents.
pipeline_8 = [{'$group': {'_id': None, 'average_age': {'$avg': '$age'}}}]

# Show results
for result in people.aggregate(pipeline_8):
    print(result)
```

### Q9: Calculate the average age of people who have tennis as their favourite sport.

```python
# Use $match to filter results before grouping so only documents
# where favourite_sports contains 'Tennis' are selected.
# Use $avg to calculate the avaerage age of the results.
pipeline_9 = [
    {'$match': {'favourite_sports': 'Tennis'}},
    {'$group': {'_id': None, 'average_age': {'$avg': '$age'}}},
]

# Show Results
for result in people.aggregate(pipeline_9):
    print(result)
```

### Q10: How many unique cities are listed in this collection?

```python
# Use distinct() to return a list of unique values for cities.
# Use dot notation, 'address.city' to access the city value from address.
# len() counts how many unique cities are in that list.
unique_cities = people.distinct('address.city')

# Show Results
print(unique_cities)
print(f'\nThere are {len(unique_cities)} unique cities in this collection.')
```

```python
# Alternative with MongoDB query only:
pipeline_10 = [
    {'$group': {'_id': '$address.city'}},
    {'$count': 'unique_cities'}
]

# Show Results
for result in people.aggregate(pipeline_10):
    print(result)
```

## Task 3: Mflix Database (PyMongo)


### Select the Mflix Sample Database

Load the sample database from MongoDB.

```python
# Select the sample_mflix database and the movies collection.
db = client.sample_mflix
movies = db.movies

# Confirm the collection is accessible by getting one document.
print(movies.find_one({}, {'_id': 0}))
```

### Q1: List the titles of all movies released after 2005 that belong to the "Thriller" genre and have an IMDb rating of 8.0 or higher. (Use projection to show only the title).

```python
# Three conditions are applied, use $gt for greater than
# Limit genre to 'Thriller' and $gte for rating equal to 8.0 or greater.
# Projection shows only the title. _id is excluded for cleaner output.
results_1 = movies.find(
    {
        'year': {'$gt': 2005},
        'genres': 'Thriller',
        'imdb.rating': {'$gte': 8.0}
    },
    {'title': 1, '_id': 0}
)

# Show results
for movie in results_1:
    print(movie)
```

### Q2: List all movies that mention the word "astronaut" (case-insensitive) in the plot.

```python
# Use $regex to match patterns in MongoDB.
# 'astronaut' matches any plot with the word 'astronaut'.
# The 'i' flag makes the search case-insensitive.
# I have chosen to limit to just the title, and
# sort them alphabetically for clarity.
results_2 = movies.find(
    {'plot': {'$regex': 'astronaut', '$options': 'i'}},
    {'title': 1, '_id': 0}
).sort('title', 1)

# Show results
for movie in results_2:
    print(movie['title'])
```
