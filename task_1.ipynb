### Q5: How many films are listed in this database?

```python
# COUNT(*) counts every row in the film table.
# Each row represents one film, so this returns the total number of films in the database.
!mysql -D sakila -e 'SELECT COUNT(*) AS "Total Films" FROM film'
```

### Q6: Display the first and last name of each actor in a single column in upper case. Name the column "Actor Name".

```python
# Examine the first 10 rows of the actor table to see its structure.
!mysql -D sakila -e 'SELECT * FROM actor LIMIT 10'
```

```python
# Q6: CONCAT joins first_name and last_name into one string with a space.
# UPPER converts the text to uppercase. The table appears to already be
# be formatted like this, but this will ensure it.
# AS 'Actor Name' renames the output column.
!mysql -D sakila -e ' SELECT UPPER(CONCAT(first_name, " ", last_name)) \
AS "Actor Name" FROM actor ORDER BY first_name'
```

### Q7: Find the ID number, first name, and last name of an actor/actors with the first name "James".

```python
# We select the three columns: actor_id, first_name, last_name.
# WHERE first_name = 'James' filters to only rows with that exact first name.
!mysql -D sakila -e 'SELECT actor_id, first_name, last_name FROM actor \
WHERE first_name = "James"'
```

### Q8: Find all actors whose last names contain the letter E. This time, order the rows by last name and first name, in that order.

```python
# LIKE matches the selected string in the chosen column, here last_name,
# ORDER BY last_name, first_name sorts by last first, then first name.
!mysql -D sakila -e 'SELECT * FROM actor WHERE last_name LIKE "%E%" \
OR last_name LIKE "%e%" ORDER BY last_name, first_name'
```

### Q9: Display the `country_id` and `country` columns for Poland, Angola, and Zambia.

```python
# Examine the first 10 rows of the country table to see its structure.
!mysql -D sakila -e 'SELECT * FROM country LIMIT 10'
```

```python
# SELECT the columns for the id and country name.
# IN ("Poland", "Angola", "Zambia") filters for multiple countries.
!mysql -D sakila -e 'SELECT country_id, country FROM country \
WHERE country IN ("Poland", "Angola", "Zambia")'
```
