### Download and Load Database
The database is downloaded, and loaded into MySQL.

```python
# Download the Sakila database archive from MySQL downloads.
!wget -q https://downloads.mysql.com/docs/sakila-db.tar.gz
```

```python
# Extract the archive containing sakila-schema.sql and sakila-data.sql.
!tar -xzf sakila-db.tar.gz
!ls sakila-db/
```

```python
# Load the schema, then load the data for the database.
!mysql -e "SOURCE sakila-db/sakila-schema.sql"
!mysql -e "SOURCE sakila-db/sakila-data.sql"
```

```python
# Verify the sakila database exists and its tables are present.
!mysql -e "SHOW DATABASES"
!mysql -D sakila -e "SHOW TABLES"
```
